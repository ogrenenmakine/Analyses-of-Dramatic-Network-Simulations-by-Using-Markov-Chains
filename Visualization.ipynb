```python
import numpy as np
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
```

```python
inter = 500
```

```python
def decompose_fft(data: list, threshold: float = 0):
    fft3 = np.fft.fft(data)
    x = np.arange(0, 5, 5 / len(data))
    freqs = np.fft.fftfreq(len(x), .01)
    recomb = np.zeros((len(x),))
    freq = []
    fft = []
    for i in range(len(fft3)):
        if abs(fft3[i]) / len(x) > threshold and abs(freqs[i]) < 1:
            sinewave = (
                1 
                / len(x) 
                * (
                    fft3[i].real 
                    * np.cos(freqs[i] * 2 * np.pi * x) 
                    - fft3[i].imag 
                    * np.sin(freqs[i] * 2 * np.pi * x)))
            recomb += sinewave
    return recomb
```

```python
y = np.array([29.1541,
39.6684,
69.1423,
46.3662,
19.2247,
42.0092,
38.4142])
y_max = y.max()
y_min = y.min()
x = np.array(list(range(len(y))))
```

```python
X_Y_Spline = make_interp_spline(x, y)
X_ = np.linspace(x.min(), x.max(), inter)
Y_ = X_Y_Spline(X_)
Y_ = (Y_-min(Y_))/(max(Y_)-min(Y_))
y = (y-min(y))/(max(y)-min(y))

recomb = decompose_fft(Y_,0.05)

# Plotting the Graph
plt.plot(x, y*(y_max-y_min)+y_min)
plt.plot(X_, recomb*(y_max-y_min)+y_min)
plt.title("Antigone:Approximation to Original Signal")
plt.xlabel("alpha")
plt.ylabel("scene number")
plt.show()
```
